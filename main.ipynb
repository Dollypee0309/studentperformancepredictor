```python
import pandas  as pd
```

Loading the data


```python
df = pd.read_csv('student_management_dataset.csv')
df.head()
```

Data Cleaning

```python
df.isnull().sum()
```

Feature Engineering

```python
print(df.columns)
```

```python
y = df['Dropout_Risk']
```

```python
x = df.drop(columns=['Student_ID','Dropout_Risk'])
x.head()
```

```python
from sklearn.preprocessing import LabelEncoder
x['Gender'] = LabelEncoder().fit_transform(df['Gender'])
x['Ethnicity'] = LabelEncoder().fit_transform(df['Ethnicity'])
x['Socioeconomic_Status'] = LabelEncoder().fit_transform(df['Socioeconomic_Status'])
x['Parental_Education_Level'] = LabelEncoder().fit_transform(df['Parental_Education_Level'])
x['Past_Academic_Performance'] = LabelEncoder().fit_transform(df['Past_Academic_Performance'])
x.head()
```

Data Normalization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

Data Splitting

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=77, stratify = y)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

MODEL BUILDING (Random Forest)

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=77)
rf_model.fit(x_train, y_train)
```

Prediction and Evaluation

```python
rf_pred = rf_model.predict(x_test)
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, f1_score, precision_score, recall_score, roc_curve, roc_auc_score

print("Random Forest Evaluation")

rf_pred = rf_model.predict(x_test)
y_proba = rf_model.predict_proba(x_test)[:, 1]

precision = precision_score(y_test, rf_pred)
recall = recall_score(y_test, rf_pred)
f1 = f1_score(y_test, rf_pred)

print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

cm = confusion_matrix(y_test, rf_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

fpr, tpr, thresholds = roc_curve(y_test, y_proba)
auc_score = roc_auc_score(y_test, y_proba)
print("AUC Score:", auc_score)

plt.figure()
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {auc_score:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # diagonal line
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate (Recall)')
plt.legend(loc='lower right')
plt.grid()
plt.show()

print(classification_report(y_test, rf_pred))
print("Accuracy:", accuracy_score(y_test, rf_pred))
```

Saving the trained model

```python
import joblib
joblib.dump(rf_model, "student_model.pkl")
```
